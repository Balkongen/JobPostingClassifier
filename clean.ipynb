```python
import pandas as pd
import json
import re

LABEL = "Bad"
FILE_PATH = "raw_data/response_1691670224081B.json"

with open(FILE_PATH) as file:
    data = json.load(file)

descs = []
for x in data["hits"]:
    posting = x["description"]["text"]
    if posting == None:
        continue
    s = posting.lower()
    # s = re.sub(r'[0-9.,\n\U0001F603&#58-]', '', s)
    # Remove urls
    # s = re.sub(r'(www\S+)|(http\S+)|(https\S+)', '', s) 
    s = re.sub('[^A-Öa-ö]+', ' ', s)
    descs.append(s)

df = pd.DataFrame(data=descs, columns=["Posting"])
df["Label"] = LABEL

df.drop_duplicates(subset=["Posting"])

display(df)

# df.to_csv("postings.csv", index=False)
df.to_csv("postings.csv", mode='a', index=False, header=False)
```
